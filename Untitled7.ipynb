Generate a histogram of `median_house_value`, create a scatter plot of `longitude` vs. `latitude` colored by `median_house_value`, calculate and display the correlation matrix, and create scatter plots to explore relationships between `median_house_value` and other features.



```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.histplot(df['median_house_value'], bins=50, kde=True)
plt.title('Distribution of Median House Values')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='median_house_value', palette='viridis', alpha=0.6)
plt.title('Geographical Distribution of House Values')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()


correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Housing Data')
plt.show()


features_to_plot = ['median_income', 'total_rooms', 'housing_median_age', 'population', 'households']
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))
axes = axes.flatten()

for i, col in enumerate(features_to_plot):
    sns.scatterplot(data=df, x=col, y='median_house_value', ax=axes[i], alpha=0.6)
    axes[i].set_title(f'Median House Value vs. {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Median House Value')


for j in range(len(features_to_plot), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous command failed again due to the mismatch between the number of income bin edges and labels. I need to correctly generate the labels to match the bins. I will retry the income binning and then proceed with calculating rooms per household and visualizing its relationship with median house value, and finally investigate spatial patterns in residuals after a linear regression.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm


income_bins = np.arange(0, df['median_income'].max() + 1, 1)
income_labels = [f'{i}-{i+1}' for i in range(0, len(income_bins) - 1)]

df['income_range'] = pd.cut(df['median_income'], bins=income_bins, labels=income_labels, right=False)

average_house_value_by_income = df.groupby('income_range', observed=False)['median_house_value'].mean().reset_index()
display(average_house_value_by_income)


df['rooms_per_household'] = df['total_rooms'] / df['households']


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='rooms_per_household', y='median_house_value', alpha=0.6)
plt.title('Median House Value vs. Rooms per Household')
plt.xlabel('Rooms per Household')
plt.ylabel('Median House Value')
plt.show()


X = df['median_income']
y = df['median_house_value']
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
df['residuals'] = model.resid


plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='residuals', palette='coolwarm', alpha=0.6)
plt.title('Spatial Distribution of Residuals (Median House Value vs. Median Income)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
